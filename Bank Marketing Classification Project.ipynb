#### When looking at the documentation for the dataset, we see that the "pdays" is: number of days that passed by after the client was last contacted from a previous campaign (numeric; -1 means client was not previously contacted). And "previous" is: number of contacts performed before this campaign and for this client.

Following this logic we can determine that if "pdays" = -1 and "previous" = 0, it means that the client has not been contacted before.

```python
# Reclassify 'poutcome' as 'nonexistent' where no previous contact occurred, and classify the rest unknowns as 'other'
X_train.loc[(X_train['pdays'] == -1) & (X_train['previous'] == 0), 'poutcome'] = 'nonexistent'
X_train['poutcome'] = X_train['poutcome'].replace('unknown', 'other')
```

# Label Encoding

```python
from sklearn.preprocessing import OneHotEncoder

# Define the categorical columns
categorical_cols = ['job', 'marital', 'education', 'month', 'poutcome', 'contact']

# Create a OneHotEncoder object with appropriate settings to handle new data
ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

# Fit and transform the training data
X_train_encoded = pd.DataFrame(ohe.fit_transform(X_train[categorical_cols]),
                               columns=ohe.get_feature_names_out(categorical_cols),
                               index=X_train.index)

# Transform the test data using the same encoder
X_test_encoded = pd.DataFrame(ohe.transform(X_test[categorical_cols]),
                              columns=ohe.get_feature_names_out(categorical_cols),
                              index=X_test.index)

# Drop original categorical columns and join encoded columns
X_train = X_train.drop(categorical_cols, axis=1).join(X_train_encoded)
X_test = X_test.drop(categorical_cols, axis=1).join(X_test_encoded)
```

```python
# Label encoding for binary columns
binary_cols = ['default', 'housing', 'loan']
le = LabelEncoder()

for col in binary_cols:
    X_train[col] = le.fit_transform(X_train[col])
    X_test[col]=le.fit_transform(X_test[col])
```

```python
from sklearn.preprocessing import MinMaxScaler

# Assuming you already encoded categorical features with pd.get_dummies
# Identify the numerical columns from the original dataset
numerical_cols = X_train.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform the training data for only numerical columns
X_train[numerical_cols] = scaler.fit_transform(X_train[numerical_cols])

# Transform the test data using the same scaler
X_test[numerical_cols] = scaler.transform(X_test[numerical_cols])
```

```python
# Lets change the true and false values of categorical features to 0 and 1. 
# This ensures consistency, as these models expect numerical values.

X_train = X_train.applymap(lambda x: 1 if x is True else (0 if x is False else x))
X_test = X_test.applymap(lambda x: 1 if x is True else (0 if x is False else x))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import math

# Function to plot percentage distribution of each column in a grid layout
def plot_percentage_distribution_grid(df, columns_per_row=4):
    total_columns = len(df.columns)
    rows = math.ceil(total_columns / columns_per_row)
    
    fig, axes = plt.subplots(rows, columns_per_row, figsize=(18, rows * 4))
    axes = axes.flatten()  # Flatten the axes array for easy indexing
    
    for i, column in enumerate(df.columns):
        # For categorical columns
        if df[column].dtype == 'object' or len(df[column].unique()) < 10:
            value_counts = df[column].value_counts(normalize=True) * 100
            value_counts.plot(kind='bar', ax=axes[i])
            axes[i].set_title(f'{column}')
            axes[i].set_xlabel('')
            axes[i].set_ylabel('Percentage (%)')
            axes[i].tick_params(axis='x', rotation=45)
        
        # For numerical columns, we use quartiles
        else:
            bins = pd.qcut(df[column], q=4, duplicates='drop')
            bin_counts = bins.value_counts(normalize=True) * 100
            bin_counts.sort_index().plot(kind='bar', ax=axes[i])
            axes[i].set_title(f'{column} (Quartiles)')
            axes[i].set_xlabel('')
            axes[i].set_ylabel('Percentage (%)')
            axes[i].tick_params(axis='x', rotation=45)
    
    # Remove any empty subplots
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])
        
    plt.tight_layout()
    plt.show()

# Call the function to plot distributions for each column in X_train
plot_percentage_distribution_grid(X_train, columns_per_row=4)
```

### Step 2.1 LogisticRegression Model


#### *** Training the model ***

```python
from sklearn.linear_model import LogisticRegression

# Initialize the Logistic Regression model
model = LogisticRegression()

# Convert target variable to numeric (0 for 'no', 1 for 'yes')
y_train = y_train.map({'no': 0, 'yes': 1})
y_test = y_test.map({'no': 0, 'yes': 1})

# Fit the model on your training data
model.fit(X_train, y_train)  # 'y_train' is the target variable in our training set
```

```python
# Predict the target values for the test dataset
y_pred = model.predict(X_test)

# Get the probability of the predictions
y_probs = model.predict_proba(X_test)[:, 1]  # Probability estimates for the positive class
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# 'yes' is the positive class and correctly specified
precision = precision_score(y_test, y_pred, pos_label=1)
recall = recall_score(y_test, y_pred, pos_label=1)
f1 = f1_score(y_test, y_pred, pos_label=1)
roc_auc = roc_auc_score(y_test, y_probs)  # Since y_probs already corresponds to 'yes'
accuracy = accuracy_score(y_test, y_pred)

print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
print(f"ROC AUC Score: {roc_auc:.2f}")
print(f"Accuracy: {accuracy:.2f}")
```
